## Compare subway traffic by day of the week. Display this as a graph.

```python
#Your code here
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df['day_of_the_week'] = df['date'].dt.day_name()
```

```python
df.head()
```

```python
df.groupby('day_of_the_week')['entries'].mean().plot(kind = 'barh')
```

```python
plt.style.use('ggplot')
day_order = ['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday']
df.groupby('day_of_the_week')['entries'].mean().loc[day_order].plot(kind = 'barh', figsize = (8, 6))
plt.ylabel('')
plt.xlabel('Number of Entries', fontsize = 14)
plt.title('Number of Entriesby Day of the Week', fontsize = 16)
plt.savefig('subway_plot')
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
#Your code here
def is_weekend(column):
    if (column == 'Saturday') | (column == 'Sunday'):
        return 'Weekend'
    else:
        return 'Not Weekend'
```

```python
df['weekend'] = df['day_of_the_week'].map(is_weekend)
```

```python
df.head()
```

```python
df.groupby('weekend')['entries'].mean()
```

```python
my_plot = df.groupby('weekend')['entries'].mean().plot(kind = 'barh', figsize = (8,6))
plt.ylabel('')
plt.xlabel ('Number Of Entries')
plt.title ('Number of Entries: Weekend vs Not Weekend')
```
